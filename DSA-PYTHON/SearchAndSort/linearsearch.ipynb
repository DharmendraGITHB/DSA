## Linear Search Algorithm
Linear search is also known as sequential and exhaustive search algorithm method. Where we start to find an element from one end to last end. Check all the elements one by one or iterate sequentially until element found and return the position of the element in the given collection.




### Algorithm: linear search
#### ncert 

LinearSearch(list, item, len(list)

Step 1: SET index = 0

Step 2: WHILE index < n, REPEAT Step 3

Step 3: IF numlist[index]= key THEN

 PRINT Element found at position, index+1
 
 STOP
 
 ELSE
 
 index = index+1
 
Step 4: PRINT Search unsuccessful


```python
#for loop
def linearsearch(arr, item):
    for i in range(len(arr)):
        if arr[i] == item:
            return i
    return None

arr= [2,4,1,6,5,8,10,32]
item = 8

func = linearsearch(arr,item)

print("element found at index", func)
    
```

```python
# while loop

def linearSearch(array, x):
    index=0
    
    while index < len(array):
        if array[index] == x:
            return index
        else:
            index = index+1
    return None

array = [3,5,1,2,9,8,10,11,12]
x= 2

a= linearSearch(array, x)

print("element found at index", a)
```

```python
#linear search using recursion

def recursionSearch(arr1, InitIndex, lastIndex, item):
    if lastIndex < InitIndex:
        return None
    
    if arr1[InitIndex] == item:
        return InitIndex
    
    if arr1[lastIndex] == item:
        return lastIndex
    
    return recursionSearch(arr1, InitIndex+1, lastIndex-1,item)           

arr1 = [2,3,7,9,1,10,4,6,5]
item = 1
InitIndex = 0
n = len(arr1)
lastIndex = n - 1

index = recursionSearch(arr1, InitIndex, lastIndex , item)

if index is None:
    print('not found')
else:
    print('found element at index', index)
```
